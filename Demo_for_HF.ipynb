```python
generator = pipeline("text-generation", model = "gpt2")
result = generator("Once upon a time in Bangalore,",
                   max_length = 50,
                   num_return_sequences = 1)
print("Generated Text: \n", result[0]['generated_text'])
```

```python
sentiment_analyzer = pipeline("sentiment-analysis")
text = "I  love learning AI with Hugging Face!"
result = sentiment_analyzer(text)
print(f" input: {text}")
print(" Sentiment:", result)
```

```python
classifier = pipeline("zero-shot-classification", model = "facebook/bart-large-mnli")
text = "This is a course on Hugging Face LLMS."
labels = ["education", "sports", "politics"]
result = classifier(text, candidate_labels = labels)
print("Input:", text)
print("Classification:", result)
```

```python
summarizer = pipeline("summarization")
article = """Hugging Face is an open-source AI company.
It provides access to pre-trained models, datasets, and tools for NLP, CV, AND speech tasks.
It is widely used by researchers and Developers."""
summary = summarizer(article, max_length = 50, min_length = 10, do_sample = False)
print("Original:", article)
print("summary:", summary[0]['summary_text'])
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification, Trainer, TrainingArguments
from datasets import load_dataset
dataset = load_dataset("imdb", split = "train[:1%]")
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")

def tokenize(batch):
  return tokenizer(batch["text"], padding = True, truncation = True)

dataset = dataset.map(tokenize, batched = True)
model = AutoModelForSequenceClassification.from_pretrained("distilbert-base-uncased")

training_args = TrainingArguments(
    output_dir = "./results",
    num_train_epochs = 1,
    per_device_train_batch_size = 2,
    logging_dir = "./logs",
    logging_steps =5
)

trainer = Trainer(model = model, args = training_args, train_dataset = dataset)
print("Fine-tuning setup ready (not training to save time).")
```
